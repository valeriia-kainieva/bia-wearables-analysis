```python
import pandas as pd
import numpy as np
data = pd.read_csv("data.csv")
data.head()
```

```python
import matplotlib.pyplot as plt

x = data["depth"]
y = data["vo2max"]

plt.figure(figsize=(6,6))

# scatter
plt.scatter(x, y, color="green", label="Data")

# линия тренда
z = np.polyfit(x, y, 1)        # 1 = линейная
p = np.poly1d(z)

plt.plot(x, p(x), color="black", label="Trend line")

plt.xlabel("Diving Depth")
plt.ylabel("Vo2max")
plt.title("Diving vs Vo2max")

plt.legend()
plt.grid()

plt.show()
```

```python
x = data["depth"]
y = data["vo2max"]

r = np.corrcoef(x,y)[0,1]

print("Correlation coefficient (r):", round(r,2))
```

A moderate negative correlation (r = -0.57) is observed between the variables; however, this is a small illustrative dataset (n = 4), so conclusions are limited.

```python
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(6,6))

sns.regplot(
    x=data["depth"],
    y=data["vo2max"],
    ci=None,
    scatter_kws={"color": "green"},
    line_kws={"color": "orange"}
)

plt.title("Depth vs Cardiovascular Endurance")
plt.xlabel("Diving Depth")
plt.ylabel("Cardiovascular Endurance")

plt.show()
```

```python
x = data["depth"]
y = data["vo2max"]

r = np.corrcoef(x, y)[0,1]

plt.figure(figsize=(6,6))

sns.regplot(x=x, y=y, ci=None, scatter_kws={"color": "green"}, line_kws={"color": "orange"})

plt.text(min(x), max(y), f"r = {r:.2f}", fontsize=12)

plt.title("Depth vs Cardiovascular Endurance")
plt.xlabel("Diving Depth")
plt.ylabel("Cardiovascular Endurance")

plt.show()
```

```python
import plotly.graph_objects as go

# данные
x = data["depth"]
y = data["vo2max"]

# корреляция
r = np.corrcoef(x, y)[0,1]

# линия тренда
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

fig = go.Figure()

# точки
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='markers',
    marker=dict(color='green', size=10),
    name="Data"
))

# линия тренда
fig.add_trace(go.Scatter(
    x=x,
    y=p(x),
    mode='lines',
    line=dict(color='black'),
    name="Trend"
))

# оформление
fig.update_layout(
    title=f"Depth vs Cardiovascular Endurance (r = {r:.2f})",
    xaxis_title="Diving Depth",
    yaxis_title="Cardiovascular Endurance"
)

fig.show()
```
